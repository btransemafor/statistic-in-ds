```python
import numpy as np 
import matplotlib.pyplot as plt 
import scipy.stats as stats
```

**Assume The data population follow poisson Distribution**

```python
# Thiet lap tham so lambda 
lambda_param = 4 
# Tao du lieu mau voi kich thuoc 1000 
population_size = 10000
np.random.seed(23)
poisson_population = np.random.poisson(lambda_param, population_size )

# In ra 10 gia tri dau tien cua mau 
print(poisson_population[:10])
```

### **Visualization Data Population (Poisson)**

```python
plt.hist(poisson_population, bins = 30, color='blue' , alpha = 0.7)
plt.title("Bieu do hist cua phan phoi POISSION")
plt.xlabel('Gia Tri')
plt.ylabel('Tan Suat')
plt.show()
```

### **Test Knowledge "The Central Limit Theorem"**

The sampling Distribution of the mean is approximately normally distributed ( the sample size >= 30)

**Step 1: Lấy mẫu ngẫu nhiên 100 giá trị từ tổng thể lần 1** 

```python
all_mean_sample = []
```

Test 

**Step 2: Lặp lại lấy mẫu ngẫu nhiên từ tổng thể với thêm ít nhất 29 lần nữa**

```python
num_samples = 1000
sample_size = 1000
for i in range(num_samples):
    random_sample = np.random.choice(poisson_population, sample_size, replace=False)
    sample_mean = np.mean(random_sample)
    all_mean_sample.append(sample_mean)
```

```python
all_mean_sample
```

```python
# Vẽ biểu đồ histogram của trung bình mẫu
plt.hist(all_mean_sample, bins=31, color='blue', alpha=0.7)
plt.title('Biểu đồ Histogram của Trung Bình Mẫu')
plt.xlabel('Giá trị Trung Bình')
plt.ylabel('Tần Suất')
plt.show()
```

```python

# Vẽ QQ plot để kiểm tra phân phối chuẩn
stats.probplot(all_mean_sample, dist="norm", plot=plt)
plt.title('QQ Plot để Kiểm Tra Phân Phối Chuẩn')
plt.show()
```
